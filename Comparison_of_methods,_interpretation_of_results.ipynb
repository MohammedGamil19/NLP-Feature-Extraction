```python
import pandas as pd

# Load the data
# Ensure this cell is run before any cell that uses 'df'
df = pd.read_csv('data.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd # Import pandas if not already imported in this cell
from sklearn.feature_extraction.text import CountVectorizer # Import CountVectorizer

# Target: category
# Load the data again here
df = pd.read_csv('data.csv')

import nltk
import re
import string
# Download stopwords if not already downloaded
try:
    stopwords = nltk.corpus.stopwords.words('english')
except LookupError:
    nltk.download('stopwords')
    stopwords = nltk.corpus.stopwords.words('english')

stop_words = set(stopwords)

def clean_text(text):
    if pd.isna(text): # Handle potential NaN values in the 'content' column
        return ""
    text = str(text).lower() # Ensure text is a string
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['cleaned'] = df['content'].apply(clean_text)


# Remove rows where 'category' is NaN *after* cleaning
df_cleaned_target = df.dropna(subset=['category'])
y = df_cleaned_target['category']

# Re-initialize and fit vectorizer_bow on the cleaned data *within this cell*
vectorizer_bow = CountVectorizer()
X_bow_cleaned_target = vectorizer_bow.fit_transform(df_cleaned_target['cleaned'])

# Train-test split using BoW features on the cleaned dat
X_train, X_test, y_train, y_test = train_test_split(X_bow_cleaned_target, y, test_size=0.2, random_state=42)

# Model training
model = LogisticRegression(max_iter=10)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# After model training and prediction
acc_bow = accuracy_score(y_test, y_pred)
f1_bow = f1_score(y_test, y_pred, average='weighted')  # 'weighted' handles multiclass

print(f"BoW - Accuracy: {acc_bow:.2f}, F1-Score: {f1_bow:.2f}")

# Results
print("BoW Results:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(df_cleaned_target['cleaned'])

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

acc_tfidf = accuracy_score(y_test, y_pred)
f1_tfidf = f1_score(y_test, y_pred, average='weighted')

print(f"TF-IDF - Accuracy: {acc_tfidf:.2f}, F1-Score: {f1_tfidf:.2f}")
print("TF-IDF Results:")
print(classification_report(y_test, y_pred))
```

```python
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import numpy as np
nltk.download('punkt')

df_cleaned_target['tokens'] = df_cleaned_target['cleaned'].apply(word_tokenize)

# Train Word2Vec
w2v_model = Word2Vec(sentences=df_cleaned_target['tokens'], vector_size=100, window=5, min_count=2, workers=4)

def document_vector(doc):
    doc = [word for word in doc if word in w2v_model.wv]
    return np.mean(w2v_model.wv[doc], axis=0) if len(doc) > 0 else np.zeros(100)

X_w2v = np.array([document_vector(doc) for doc in df_cleaned_target['tokens']])

# Remove rows with NaN vectors
valid_idx = ~np.isnan(X_w2v).any(axis=1)
X_w2v = X_w2v[valid_idx]
y_valid = y.iloc[valid_idx]

X_train, X_test, y_train, y_test = train_test_split(X_w2v, y_valid, test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=10)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

acc_w2v = accuracy_score(y_test, y_pred)
f1_w2v = f1_score(y_test, y_pred, average='weighted')

print(f"Word2Vec - Accuracy: {acc_w2v:.2f}, F1-Score: {f1_w2v:.2f}")
print("Word2Vec Results:")
print(classification_report(y_test, y_pred))
```
